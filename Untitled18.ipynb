```python
import os
import numpy as np
from PIL import Image, ImageOps
from google.colab import files
from IPython.display import display

def apply_palette(img, hex_colors):
    """Manually forces the image to use a specific set of colors."""
    # Convert hex to RGB tuples
    rgb_colors = []
    for h in hex_colors:
        h = h.lstrip('#')
        rgb_colors.extend(int(h[i:i+2], 16) for i in (0, 2, 4))

    # Pad palette to 256 colors (768 values) for PIL
    palette_data = rgb_colors + [0] * (768 - len(rgb_colors))

    # Create a palette image template
    palette_img = Image.new('P', (1, 1))
    palette_img.putpalette(palette_data)

    # Quantize the original image to this palette with dithering
    return img.convert('RGB').quantize(palette=palette_img, dither=Image.FLOYDSTEINBERG)

def process_image(input_path, style_choice):
    img = Image.open(input_path)

    # Define Styles
    styles = {
        "1": {"name": "Retro (High Dither)", "width": 140, "colors": 6, "palette": None},
        "2": {"name": "GameBoy Olive", "width": 120, "colors": 4, "palette": ["#0f380f", "#306230", "#8bac0f", "#9bbc0f"]},
        "3": {"name": "Cyberpunk", "width": 160, "colors": 5, "palette": ["#000b1e", "#ff0055", "#00ff9f", "#00b8ff", "#fdee00"]},
        "4": {"name": "Noir (B&W)", "width": 180, "colors": 3, "palette": ["#000000", "#777777", "#FFFFFF"]},
        "5": {"name": "Sunset Gold", "width": 150, "colors": 5, "palette": ["#2d1b33", "#7b3445", "#f06449", "#ede89a", "#63ace5"]}
    }

    config = styles.get(style_choice, styles["1"])
    print(f"Applying Style: {config['name']}...")

    # Resize (Downscale)
    w_percent = (config['width'] / float(img.size[0]))
    h_size = int((float(img.size[1]) * float(w_percent)))
    img_small = img.resize((config['width'], h_size), resample=Image.BICUBIC)

    # Apply Color logic
    if config['palette']:
        img_pixel = apply_palette(img_small, config['palette'])
    else:
        # Adaptive palette based on the image itself
        img_pixel = img_small.convert('P', palette=Image.ADAPTIVE, colors=config['colors'])

    # Upscale back to original size
    final_output = img_pixel.resize(img.size, resample=Image.NEAREST)
    return final_output.convert('RGB')

# --- Colab Interaction ---

uploaded = files.upload()

if uploaded:
    filename = list(uploaded.keys())[0]

    print("\nSelect a style feel:")
    print("1: Retro / Original Look (Adaptive colors)")
    print("2: GameBoy Olive (Classic handheld)")
    print("3: Cyberpunk (Neon pinks/blues)")
    print("4: Noir (High contrast monochrome)")
    print("5: Sunset Gold (Warm vintage)")

    choice = input("\nEnter style number (1-5): ")

    result = process_image(filename, choice)

    print("\nResult:")
    display(result)

    # Save result
    result.save("pixel_art_output.png")
```

```python
import os
from PIL import Image, ImageFilter, ImageOps
from google.colab import files
from IPython.display import display

def create_clean_pixel_art(input_path, style_choice):
    # 1. Open the image
    img = Image.open(input_path).convert('RGB')

    # --- THE NEW SMOOTHING LOGIC ---
    # We apply a Median Filter to flatten textures before pixelating.
    # A 'size' of 5, 7, or 9 works best. Larger = smoother/flatter.
    smoothing_strength = 7
    print(f"Applying smoothing filter (strength {smoothing_strength})...")
    img_smoothed = img.filter(ImageFilter.MedianFilter(size=smoothing_strength))
    # -------------------------------

    # Define Styles (Tweaked for cleaner looks)
    styles = {
        # Style 1: Best for that "Instagram Filter" look. Adaptive colors, but smoothed first.
        "1": {"name": "Clean Retro (Adaptive)", "width": 150, "colors": 7, "palette": None},
        # Style 2: Forces a specific warm, limited palette often seen in these edits.
        "2": {"name": "Warm Vibe (Fixed Palette)", "width": 140, "colors": 5, "palette": ["#2a1d2e", "#563645", "#8f565b", "#c98d7a", "#f3c6aa"]},
        # Style 3: Dark and moody, good for low light.
        "3": {"name": "Midnight (Fixed Palette)", "width": 160, "colors": 4, "palette": ["#0d1117", "#1f2937", "#374151", "#6b7280"]},
    }

    config = styles.get(style_choice, styles["1"])
    print(f"Applying Style: {config['name']} with {config['colors']} colors...")

    # 2. Downscale (Using BILINEAR on the smoothed image creates softer transitions)
    w_percent = (config['width'] / float(img_smoothed.size[0]))
    h_size = int((float(img_smoothed.size[1]) * float(w_percent)))
    img_small = img_smoothed.resize((config['width'], h_size), resample=Image.BILINEAR)

    # 3. Quantization & Dithering
    # Because the input is now smooth, the dithering will be structured instead of chaotic.
    if config['palette']:
        # Process fixed palette
        rgb_colors = []
        for h in config['palette']:
            h = h.lstrip('#')
            rgb_colors.extend(int(h[i:i+2], 16) for i in (0, 2, 4))
        palette_data = rgb_colors + [0] * (768 - len(rgb_colors))
        palette_img = Image.new('P', (1, 1))
        palette_img.putpalette(palette_data)
        # Use FLOYDSTEINBERG dither, which looks cleaner on smooth inputs
        img_pixel = img_small.quantize(palette=palette_img, dither=Image.FLOYDSTEINBERG)
    else:
        # Process adaptive palette
        img_pixel = img_small.convert('P', palette=Image.ADAPTIVE, colors=config['colors'], dither=Image.FLOYDSTEINBERG)

    # 4. Upscale back to original size with sharp edges
    final_output = img_pixel.resize(img.size, resample=Image.NEAREST)
    return final_output.convert('RGB')

# --- Colab Interaction ---

print("Please upload an image:")
uploaded = files.upload()

if uploaded:
    filename = list(uploaded.keys())[0]

    print("\nSelect a style feel:")
    print("1: Clean Retro (Uses image's own colors, best match for your examples)")
    print("2: Warm Vibe (Forces reddish/peach tones)")
    print("3: Midnight (Dark bluish tones)")

    choice = input("\nEnter style number (1-3): ")

    # Process filter
    result = create_clean_pixel_art(filename, choice)

    print("\nImproved Result:")
    display(result)

    # Save
    result.save("clean_pixel_art_output.png")
else:
    print("No file uploaded.")
```
